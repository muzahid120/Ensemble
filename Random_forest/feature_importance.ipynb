```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
x,y=load_digits(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import numpy as np
from skimage.transform import resize

# Assuming x[5] is your input array of size 64
original_image = x[5].reshape(8, 8)  # Reshape it to 8x8 for example

# Resize the image to 28x28
resized_image = resize(original_image, (28, 28))

# Now you can visualize the resized image
sns.heatmap(resized_image)
```

```python
plt.imshow(resized_image)
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_test,y_pred)
```
